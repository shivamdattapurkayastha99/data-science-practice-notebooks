```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,3,4]
y=[23,45,65]
x1=[3,7,3]
y1=[5,8,9]
plt.plot(x,y,label="shivam",linewidth=4)
plt.plot(x1,y1,label="baba",linewidth=3)
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,3,4]
y=[23,45,65]
x1=[3,7,3]
y1=[5,8,9]
plt.plot(x,y,label="shivam",linewidth=4)
plt.scatter(x1,y1,label="baba")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,3,4]
y=[23,45,65]
x1=[3,7,3]
y1=[5,8,9]
plt.hist(x,label="shivam",linewidth=4)
plt.scatter(x1,y1,label="baba")
plt.legend()
plt.show()
```
